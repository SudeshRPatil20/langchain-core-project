# Prompt Template Test Summerization

```python
from langchain.chains import LLMChain
from langchain import PromptTemplate

generictemplate="""
write a summery of the following speech:
Speech:{speech}
Translate the precise Summery in :{language}
"""

prompt=PromptTemplate(
    input_variables=['speech','language'],
    template=generictemplate
)

prompt
```

```python
prompt.format(speech=speech, language="Hindi")
```

```python
llm_chain=LLMChain(llm=llm, prompt=prompt)
summery=llm_chain.run({'speech':speech, 'language':'Hindi'})
summery
```

As you can see content is summerize and also translated to language but it is sufficiant for just small paragraphs but not for larger documents

# for this we use StuffDocument Chain Text Summerization

```python
from langchain_community.document_loaders import PyPDFLoader

loader=PyPDFLoader("apjspeech.pdf")
docs=loader.load_and_split()
docs
```

```python
template=""" Write a concise short summery of the following speech,
Speech:{text}

"""

prompt=PromptTemplate(input_variables=['text'],
                      template=template)
```

```python
from langchain.chains.summarize import load_summarize_chain
chain=load_summarize_chain(llm=llm, chain_type="stuff", prompt=prompt,verbose=True)
output_summery=chain.run(docs)
output_summery
```

# Map Reducer to Summerize large documents

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
docs
```

```python
final_documents=RecursiveCharacterTextSplitter(chunk_size=2000, chunk_overlap=100).split_documents(docs)
final_documents
```

```python
len(final_documents)
```

```python
chunk_prompt="""
Please summarize the below speech:
Speech:`{text}'
Summary:
"""

map_prompt_template=PromptTemplate(input_variables=['text'],
                                   template=chunk_prompt)
```

```python
final_prompt='''
Provide the final Summary of the entire points with this important points.
Add a Motivation title, Start the precise summery with an introduction and provide the summary with points for the speech.
Speech:{text}

'''

final_prompt_template=PromptTemplate(input_variables=['text'], template=final_prompt)
final_prompt_template
```

```python
summery_chain=load_summarize_chain(
    llm=llm,
    chain_type="map_reduce",
    map_prompt=map_prompt_template,
    combine_prompt=final_prompt_template,
    verbose=True
)

output=summery_chain.run(final_documents)
output
```

# Refine Chain Summerization

```python
chain=load_summarize_chain(
    llm=llm,
    chain_type="refine",
    verbose=True
)

output_summery=chain.run(final_documents)
```
